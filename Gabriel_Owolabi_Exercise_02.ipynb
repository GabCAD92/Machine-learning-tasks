```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# EXERCISE 1
# =========================

x = np.linspace(1, 1000, 500)

y1 = 10*x - 4
y2 = x**2

plt.figure(figsize=(8,6))

plt.plot(x, y1, color='red', linewidth=3, label='y=10.x-4')
plt.plot(x, y2, color='green', linewidth=3, label='y=x²')

plt.yscale('log')  # VERY IMPORTANT

plt.title('My first graph in Machine Learning class', fontsize=14)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

plt.legend(loc='lower right')
plt.grid(True, which='both', linestyle='-', alpha=0.4)

plt.show()



# =========================
# EXERCISE 2
# =========================

x = np.linspace(1, 1000, 500)

y1 = 10*x - 4
y2 = x**2

fig = plt.figure(figsize=(10,6))

# ---- LEFT: FUNCTIONS ----
ax1 = plt.subplot(1,2,1)

ax1.plot(x, y1, color='red', linewidth=3, label='y=10.x-4')
ax1.plot(x, y2, color='green', linewidth=3, label='y=x²')

ax1.set_yscale('log')
ax1.set_title('Functions:')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax1.legend()
ax1.grid(True, which='both', alpha=0.4)

# ---- RIGHT: INTERSECTION ----
ax2 = plt.subplot(1,2,2)

# zoom range
x_zoom = np.linspace(50, 150, 200)

y1_zoom = 10*x_zoom - 4
y2_zoom = x_zoom**2

ax2.plot(x_zoom, y1_zoom, color='red', linewidth=3)
ax2.plot(x_zoom, y2_zoom, color='green', linewidth=3)

ax2.set_yscale('log')
ax2.set_title('Intersection:')

# intersection approx (visually around x=100)
x_int = 100
y_int = 10*x_int - 4

# point
ax2.scatter(x_int, y_int, color='black', s=80, zorder=5)

# dashed lines
ax2.axvline(x_int, color='red', linestyle='--')
ax2.axhline(y_int, color='red', linestyle='--')

# arrow + text
ax2.annotate('Point',
             xy=(x_int, y_int),
             xytext=(110, 500),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center')

ax2.grid(True, which='both', alpha=0.4)

plt.tight_layout()
plt.show()



# =========================
# EXERCISE 3
# =========================

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('clusters.dat')

x = data[:,0]
y = data[:,1]
group = data[:,2]

fig = plt.figure(figsize=(10,8))

# ---- MAIN SCATTER ----
ax_scatter = plt.subplot2grid((2,2), (0,0), colspan=1)
scatter = ax_scatter.scatter(x, y, c=group, cmap='tab10', marker='+')

ax_scatter.set_title('My first data file:')
ax_scatter.set_xlabel('')
ax_scatter.set_ylabel('')

# ---- RIGHT HIST (Y) ----
ax_histy = plt.subplot2grid((2,2), (0,1))
ax_histy.hist(y, bins=20, orientation='horizontal')
ax_histy.set_xlabel('Occurrences')
ax_histy.set_ylabel('Y Values')

# ---- BOTTOM HIST (X) ----
ax_histx = plt.subplot2grid((2,2), (1,0))
ax_histx.hist(x, bins=20)
ax_histx.set_xlabel('X Values')
ax_histx.set_ylabel('Occurrences')

plt.tight_layout()
plt.show()
```
